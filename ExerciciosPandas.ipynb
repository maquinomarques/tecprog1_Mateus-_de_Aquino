1) Utilizando o DataFrame importado (alunos3.csv) calcule a média das provas em uma nova coluna chamada (Media_provas)

```python
df = pd.read_csv('dados/alunos3.csv', sep = ';', decimal = ',')
df
```

```python
df['Media'] = round(df.loc[ : ,['Prova_1','Prova_2','Prova_3','Prova_4']].mean(axis = 1),2)
df
```

2) Quem foram os alunos que obtiveram a maior e a menor média

```python
maior_nota = df[df['Media'] == df['Media'].max()]
maior_nota
```

```python
menos_nota = df[df['Media'] == df['Media'].min()]
menos_nota
```

7. Leia o arquivo `aluno3.csv`, apenas as colunas RA, Prova_1, Prova_2, Prova_3, Prova_4 calcule a médias das notas da prova e salve em um arquivo excel (xlsx) as colunas RA e Média.


```python
alunos3 = pd.read_csv('dados/alunos3.csv', usecols=['RA', 'Prova_1', 'Prova_2', 'Prova_3', 'Prova_4'] , sep=';', decimal = ",")
```

```python
alunos3['Media'] = round(alunos3.loc[ : ,['Prova_1','Prova_2','Prova_3','Prova_4']].mean(axis = 1),2)
```

```python
writer= pd.ExcelWriter('dados/questão_07.xlsx', mode='w')
alunos3.to_excel(writer, columns=['RA', 'Media'], index=False)
writer.close()
```
